```python
from PIL import Image
import numpy as np
from model.deformer import deform_card
from model.prediction import card_prediction_processing

def predict_card_set(
    img_path,
    model_left=model_left,
    model_right=model_right,
    le_left=le_left,
    le_right=le_right,
    show_debug=False
):
    """
    Given a path to a Pokémon card image, predicts the set symbol.
    Returns the predicted set (and optionally shows debugging info).
    """
    # 1. Load and align image
    pil_img = Image.open(img_path).convert("RGB")
    aligned_img = deform_card(pil_img)

    # 2. Crop bottom corners
    graybottomleft, graybottomright = card_prediction_processing(aligned_img)

    if show_debug:
        import matplotlib.pyplot as plt
        plt.figure(figsize=(10,4))
        plt.subplot(1,2,1)
        plt.imshow(graybottomleft[0, :, :, 0], cmap='gray')
        plt.title("Bottom Left Crop")
        plt.axis('off')
        plt.subplot(1,2,2)
        plt.imshow(graybottomright[0, :, :, 0], cmap='gray')
        plt.title("Bottom Right Crop")
        plt.axis('off')
        plt.show()

    # 3. Prepare input (already normalized!)
    X_left = graybottomleft.astype(np.float32)
    X_right = graybottomright.astype(np.float32)

    # 4. Predict
    pred_left = model_left.predict(X_left)
    pred_right = model_right.predict(X_right)

    set_left = le_left.inverse_transform([np.argmax(pred_left)])[0]
    set_right = le_right.inverse_transform([np.argmax(pred_right)])[0]
    conf_left = np.max(pred_left)
    conf_right = np.max(pred_right)

    # 5. Decision logic
    if set_left == 'no' and set_right == 'no':
        result = "No set symbol detected! Try a clearer image."
    elif set_left != 'no' and set_right != 'no':
        if conf_left >= conf_right:
            result = f"Ambiguous, both detected. Picking LEFT: {set_left} (conf {conf_left:.2f})"
        else:
            result = f"Ambiguous, both detected. Picking RIGHT: {set_right} (conf {conf_right:.2f})"
    elif set_left != 'no':
        result = f"Predicted set (left corner): {set_left} (conf {conf_left:.2f})"
    else:
        result = f"Predicted set (right corner): {set_right} (conf {conf_right:.2f})"

    print(result)
    return {
        "left": (set_left, conf_left),
        "right": (set_right, conf_right),
        "result": result
    }
```

```python
# Just call for any image!
output = predict_card_set("images_example/IMG_9401.jpg", show_debug=True)
print(output)
```

```python
from PIL import Image
import numpy as np
from model.deformer import deform_card
from model.prediction import card_prediction_processing
from model.text_detection import get_pokeid  # or wherever your get_pokeid lives
from model.prediction import get_card_info         # replace path as needed

def predict_and_fetch_card_info(
    img_path,
    model_left=model_left,
    model_right=model_right,
    le_left=le_left,
    le_right=le_right,
    show_debug=False
):
    # 1. Load & deskew card image
    pil_img = Image.open(img_path).convert("RGB")
    aligned_img = deform_card(pil_img)

    # 2. Crop bottom corners
    graybottomleft, graybottomright = card_prediction_processing(aligned_img)

    if show_debug:
        plt.figure(figsize=(10,4))
        plt.subplot(1,2,1)
        plt.imshow(graybottomleft[0, :, :, 0], cmap='gray')
        plt.title("Bottom Left Crop")
        plt.axis('off')
        plt.subplot(1,2,2)
        plt.imshow(graybottomright[0, :, :, 0], cmap='gray')
        plt.title("Bottom Right Crop")
        plt.axis('off')
        plt.show()


    # 3. Prepare input
    X_left = graybottomleft.astype(np.float32)
    X_right = graybottomright.astype(np.float32)

    # 4. Predict set
    pred_left = model_left.predict(X_left)
    pred_right = model_right.predict(X_right)
    set_left = le_left.inverse_transform([np.argmax(pred_left)])[0]
    set_right = le_right.inverse_transform([np.argmax(pred_right)])[0]
    conf_left = np.max(pred_left)
    conf_right = np.max(pred_right)

    # 5. Pick set_id
    if set_left == 'no' and set_right == 'no':
        print("No set symbol detected! Try a clearer image.")
        set_id = None
    elif set_left != 'no' and set_right != 'no':
        set_id = set_left if conf_left >= conf_right else set_right
        print(f"Ambiguous, picking {'LEFT' if conf_left >= conf_right else 'RIGHT'}: {set_id}")
    elif set_left != 'no':
        set_id = set_left
        print(f"Predicted set (left): {set_left}")
    else:
        set_id = set_right
        print(f"Predicted set (right): {set_right}")

    # 6. Get card number via OCR
    if set_id is not None:
        poke_id = get_pokeid(np.array(aligned_img), set_id)
        print(f"Extracted card number: {poke_id}")

        # 7. Fetch info from API
        print(f"Looking up info for set={set_id!r}, card={poke_id!r}…")
        info = get_card_info(set_id, poke_id)
        if info is None:
            print("⚠️  API lookup failed.")
            return None
        else:
            rarity, market_price, image_url = info
            print(f"Rarity:       {rarity}")
            print(f"Avg. Price:   ${market_price:.2f}")
            print(f"Card image:   {image_url}")
            return {
                "set_id": set_id,
                "poke_id": poke_id,
                "rarity": rarity,
                "market_price": market_price,
                "image_url": image_url
            }
    else:
        print("Set symbol was not detected, cannot proceed to OCR or API lookup.")
        return None
```

```python
import cv2
import matplotlib.pyplot as plt

def debug_edges(img):
    # img: RGB or BGR np.array
    if img.shape[2] == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    else:
        gray = img
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    edges = cv2.Canny(blur, 50, 150)
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.imshow(gray, cmap='gray')
    plt.title('Gray')
    plt.axis('off')
    plt.subplot(1,3,2)
    plt.imshow(blur, cmap='gray')
    plt.title('Blurred')
    plt.axis('off')
    plt.subplot(1,3,3)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny edges')
    plt.axis('off')
    plt.show()
    return edges
```

```python
img_path = "images_example/IMG_9416.jpg"
img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
edges = debug_edges(img_rgb)
```
